```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Tweets.csv')
```

```python
df = df[['text', 'airline_sentiment']]
```

```python
df.head()
```

```python
df.airline_sentiment.unique()
```

```python
df = df[df['airline_sentiment']!='neutral']
```

```python
sentiment_label = df['airline_sentiment'].factorize()
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tweet = df['text'].values #.values gives OP as an array
```

```python
tokenizer = Tokenizer(num_words=5000)
```

```python
tokenizer.fit_on_texts(tweet)
```

```python
tokenizer.word_index
```

```python
vocab_size = len(tokenizer.word_index)
```

```python
tokenizer.texts_to_sequences(tweet)
```

```python
encoded_data = tokenizer.texts_to_sequences(tweet)
```

```python
#pad_sequences to make reviews all of same length

padded_sequnece = pad_sequences(encoded_data, padding='pre', maxlen=200)
```

```python
print(encoded_data[0])
```

```python
print(padded_sequnece[0])
```

```python
tweet[0]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM, Embedding
```

```python
model = Sequential()

model.add(Embedding(vocab_size, 32, input_length = 200))
model.add(LSTM(50, dropout=0.5, recurrent_dropout=0.5))
model.add(Dropout(0.2))

model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='BinaryCrossentropy', metrics = ['accuracy'])
```

```python
model.summary()
```

```python
sentiment_label
```

```python
sentiment_label[0]
```

```python
model.fit(padded_sequnece, sentiment_label[0], epochs = 5, validation_split=0.2)
```

```python
text = "it is wow"

tw = tokenizer.texts_to_sequences([text])
tw = pad_sequences(tw, maxlen=200)

prediction = int(model.predict(tw).round().item())
```
